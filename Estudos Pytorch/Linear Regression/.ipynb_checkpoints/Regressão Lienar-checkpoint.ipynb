# Regressão Linear com PyTorch


A Regressão Linear é modelo matemático, que associa a variável dependente com a variável independente. A variável dependente é aquela que está sendo explicada, enquanto a variável independente é aquela que é utilizada para explicar a variação na variável dependente.

Um modelo de regressão linear é uma equação matemática que fornece uma relação linear, ou seja, de linha reta entre duas variáveis, comumente chamada de x e y:

$y = a + bx$

onde: 
* y: Variável dependente
* a: Termo constante ou intercepto de y
* b: Inclinação da reta
* x: Variável independente



Nesse exemplo iremos fazer a regressão Linear da equação:

$ y = 2x + 1$

```python
import numpy as np

# Criando um dataset simples para treino

x_values = [i for i in range(11)]
x_train = np.array(x_values, dtype=np.float32)
x_train = x_train.reshape(-1, 1)

y_values = [2*i + 1 for i in x_values]
y_train = np.array(y_values, dtype=np.float32)
y_train = y_train.reshape(-1, 1)
```

```python
# Modelo

import torch
from torch.autograd import Variable

class linearRegression(torch.nn.Module):
    def __init__(self, inputSize, outputSize):
        super(linearRegression, self).__init__()
        self.linear = torch.nn.Linear(inputSize, outputSize)

    def forward(self, x):
        out = self.linear(x)
        return out
```

```python
# Dimensoes das variaveis 'x' e 'y'
inputDim = 1         
outputDim = 1

learningRate = 0.01 
epochs = 100

model = linearRegression(inputDim, outputDim)
##### Usar GPU #######
if torch.cuda.is_available():
    model.cuda()
    
```

```python
# Loss Function
criterion = torch.nn.MSELoss() 

# Optimization Function
optimizer = torch.optim.SGD(model.parameters(), lr=learningRate)
```

```python
# Treinar modelo

for epoch in range(epochs):
    # Convertendo inputs e labels em Variable (Gradientes)
    if torch.cuda.is_available():
        inputs = Variable(torch.from_numpy(x_train).cuda())
        labels = Variable(torch.from_numpy(y_train).cuda())
    else:
        inputs = Variable(torch.from_numpy(x_train))
        labels = Variable(torch.from_numpy(y_train))

    # Limpe os buffers de gradiente porque não queremos que nenhum gradiente da época anterior seja transportado, 
    # não queremos acumular gradientes
    optimizer.zero_grad()

    # obter a saída do modelo, dadas as entradas
    outputs = model(inputs)

    # obter perda para a produção prevista
    loss = criterion(outputs, labels)
    print(loss)
    # obter gradientes para parâmetros
    loss.backward()

    # update dos parâmetros
    optimizer.step()

    print(f'epoch {epoch}, loss {loss}')
```

```python
# Prediction

with torch.no_grad(): 
    if torch.cuda.is_available():
        predicted = model(Variable(torch.from_numpy(x_train).cuda())).cpu().data.numpy()
    else:
        predicted = model(Variable(torch.from_numpy(x_train))).data.numpy()
    print(predicted)

import matplotlib.pyplot as plt
plt.clf()
plt.plot(x_train, y_train, 'go', label='Dados reais', alpha=0.5)
plt.plot(x_train, predicted, '--', label='Predictions', alpha=0.5)
plt.legend(loc='best')
plt.show()
```

```python
# Caminho do save dos pesos
save_model = "./models/weigths.pt"

# Armazenar em um state_dict os arquivo
torch.save(model, save_model)

# definir modelo: pesos sao iniciados randomicos
_model = linearRegression(inputDim, outputDim)

weights = torch.load(save_model)

_model=torch.load(save_model)
```
